```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
df = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
df
```

```python
df['label'] = breast_cancer_dataset.target
df
```

```python
x=df.drop('label',axis=1)
y=df['label']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
print(x_train)
```

```python
x_test=scaler.fit_transform(x_test)
print(x_test)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), #flatten- to convert 1 dimensional arr, 30 columns,features
    keras.layers.Dense(20, activation='relu'),#dense- hidden layer,here 20 hidden layer
    keras.layers.Dense(2,activation='sigmoid')#2 output layer
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, validation_split=0.1, epochs=10)
```

```python
history = model.fit(x_train, y_train, validation_split=0.1, epochs=50)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation'],loc= 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation'],loc= 'lower right')
```
